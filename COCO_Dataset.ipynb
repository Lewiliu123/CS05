````python
import ijson
import pandas as pd

# Path to COCO JSON file
json_path = "instances_train2017.json"

# Get JSON Top-Level Keys
print("Getting JSON Top-Level Keys...")
with open(json_path, "r", encoding="utf-8") as f:
    parser = ijson.parse(f)
    top_keys = set()
    for prefix, event, value in parser:
        if prefix not in top_keys:
            print(f"Found top-level key: {prefix}")
            top_keys.add(prefix)
        # Display only the first 20 keys
        if len(top_keys) >= 20:
            break
````

````python
# Extract 'images' Information
print("\n Extracting Image Information...")
with open(json_path, "r", encoding="utf-8") as f:
    images = ijson.items(f, "images.item")
    for i, img in enumerate(images):
        print(f"Image ID: {img['id']}, Filename: {img['file_name']}, Width: {img['width']}, Height: {img['height']}")
        # Show only first 5 images
        if i >= 5:
            break
````

````python
# Extract 'annotations' Information
print("\n Extracting Annotation Information...")
with open(json_path, "r", encoding="utf-8") as f:
    annotations = ijson.items(f, "annotations.item")
    for i, ann in enumerate(annotations):
        print(f"Image ID: {ann['image_id']}, Category ID: {ann['category_id']}, Bounding Box: {ann['bbox']}")
        if i >= 5:  # Show only first 5 annotations
            break
````

````python
# Extract 'categories' Information
print("\n Extracting Categories Information...")
with open(json_path, "r", encoding="utf-8") as f:
    categories = ijson.items(f, "categories.item")
    for cat in categories:
        print(f"Category ID: {cat['id']}, Category Name: {cat['name']}")
````

````python
# Extract 'categories' Information
print("\n Extracting Categories Information...")
categories_list = []
with open(json_path, "r", encoding="utf-8") as f:
    categories = ijson.items(f, "categories.item")
    for cat in categories:
        categories_list.append({"Category ID": cat["id"], "Category Name": cat["name"]})

# To DataFrame
df_categories = pd.DataFrame(categories_list)
df_categories
````

````python
# Filter Ocean-Related Categories (boat, surfboard, fish, bird)
ocean_categories = {"boat", "surfboard", "bird", "skateboard"}

print("\n Filtering Ocean-Related Categories...")
filtered_annotations = []
filtered_images = set()

with open(json_path, "r", encoding="utf-8") as f:
    # Get category mapping
    category_map = {cat["id"]: cat["name"] for cat in ijson.items(f, "categories.item")}
    ocean_category_ids = {k for k, v in category_map.items() if v in ocean_categories}

with open(json_path, "r", encoding="utf-8") as f:
    # Extract annotations related to ocean objects
    for ann in ijson.items(f, "annotations.item"):
        if ann["category_id"] in ocean_category_ids:
            filtered_annotations.append(ann)
            filtered_images.add(ann["image_id"])
        if len(filtered_annotations) >= 5:  # Show only first 5 ocean-related annotations
            break

print(f"Found {len(filtered_annotations)} annotations containing ocean-related objects.")
````
